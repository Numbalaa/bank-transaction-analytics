Data Cleaning:
fixing the data types for dates

```python
#converting the data columns from text to actual dates

df['TransactionDate'] = pd.to_datetime(df['TransactionDate'])
df['PreviousTransactionDate'] = pd.to_datetime(df['PreviousTransactionDate'])
```

```python
#confirming to see if it worked

print(df.dtypes)
```

Extraction useful information

```python
# creating new column to
df['Month'] = df['TransactionDate'].dt.month
#
df['DayOfWeek'] = df['TransactionDate'].dt.strftime('%A')
#
df['Hour'] = df['TransactionDate'].dt.hour
#checking to see if it worked

print(df[['TransactionDate', 'Month', 'DayOfWeek', 'Hour']].head())

```

Analysis Time!

```python
#how many transactions of each type?
print("Transaction Types:")
print(df['TransactionType'].value_counts())

#which channel is most used?
print("\nChannels:")
print(df['Channel'].value_counts())

#what is the avg transaction amount?
print("\nAverage Transaction Amount:")
print(df['TransactionAmount'].mean())

#What is the busiest month?
print("\nTransaction per Month:")
print(df['Month'].value_counts().sort_index())
```
